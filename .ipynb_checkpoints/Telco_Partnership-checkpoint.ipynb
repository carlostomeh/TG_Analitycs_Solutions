## 1.0 Imports

```python
import pandas as pd
import os
from scipy import stats
from scipy.stats import norm
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Pre processamento
from sklearn.base import BaseEstimator
from sklearn.base import TransformerMixin
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import KFold

# Modelagem e escolha de parametros
from sklearn.linear_model import LogisticRegression
#from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# Métricas
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from sklearn.metrics import plot_confusion_matrix

from sklearn import tree

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

## 2.0 EDA

### Metadados

O arquivo contém **7043 linhas** (clientes) e **21 colunas** (atributos), descritos abaixo:

### Atributos:

- CustomerID: ID do cliente **(char)**
- gender: Gênero do cliente ('Male' ou 'Female') **(char)**
- SeniorCitizen: Se o cliente é idoso ou não (1, 0) **(boolean: 0/1)**
- Partner: Se o cliente tem parceiro ou não (Yes, No) **(boolean: y/n)**
- Dependents: Se o cliente tem dependentes ou não (Yes, No) **(boolean: y/n)**
- tenure: Número de meses que o cliente tem relacionamento com a empresa **(integer)**
- PhoneService: Se o cliente tem um serviço de telefone ou não (Yes, No) **(**boolean: y/n**)**
- MultipleLines: Se o cliente tem múltiplas linhas ou não (Yes, No, No phone service) **(boolean: y/n)**
- InternetService: Serviço de Provedor de Internet do cliente (DSL, Fiber optic, No) **(categorical)**
- OnlineSecurity: Se o cliente tem serviço de segurança online ou não (Yes, No, No internet service) **(boolean: y/n)**
- OnlineBackup: Se o cliente tem serviço de backup online ou não (Yes, No, No internet service) **(boolean: y/n)**
- DeviceProtection: Se o cliente tem serviço de proteção de dispositivo ou não (Yes, No, No internet service) **(boolean: y/n)**
- TechSupport: Se o cliente tem suporte técnico ou não (Yes, No, No internet service) **(boolean: y/n)**
- StreamingTV: Se o cliente tem serviço de streaming de TV ou não (Yes, No, No internet service) **(boolean: y/n)**
- StreamingMovies: Se o cliente tem serviço de streaming de filmes ou não (Yes, No, No internet service) **(boolean: y/n)**
- Contract: Os termos do contrato do cliente: Mensal, Anual ou Bianual (Month-to-month, One year, Two year) **(categorical)**
- PaperlessBilling: Se o cliente tem fatura digital (não impressa) ou não (Yes, No) **(boolean: y/n)**
- PaymentMethod: O método de pagamento do cliente (Electronic check, Mailed check, Bank transfer (automatic), Credit card (automatic)) **(categorical)**
- MonthlyCharges: O total cobrado do cliente mensalmente **(real number)**
- TotalCharges: As cobranças totais do cliente **(real number)**
- Churn: Se o cliente abandonou os serviços da impresa ou não, nossa variável Target (Yes or No) **(boolean: y/n)**

```python
path = 'datasets/Telco-Customer-Churn.csv'
df_telco = pd.read_csv(path)
df_telco.head()
```

```python
df_telco.info()
```

```python
print('Este conjunto de dados possui ', df_telco.shape[1],' colunas e ', df_telco.shape[0], ' linhas. Dessas estão representados', len(set(df_telco.customerID))
          , 'clientes.\n')
```

## 3.0 Data Processing

```python
# Uma vez que TotalCharges é um object, vamos mudar seu tipo para float
df_telco["TotalCharges"] = pd.to_numeric(df_telco.TotalCharges, errors='coerce')

# Temos apenas 11 missing values para TotalCharges, então vamos dropar as linhas que contém esses missings.
df_telco = df_telco.dropna()

# Transformando a variável resposta em booleana
df_telco['Churn'].replace(to_replace='Yes', value=1, inplace=True)
df_telco['Churn'].replace(to_replace='No',  value=0, inplace=True)

# Vamos criar variáveis dummy para os nossos atributos (e retirar a coluna do customerID, que não será utilizada)
df_telco = pd.get_dummies(df_telco.drop("customerID",axis=1),drop_first=True)


```

```python
fig, axes = plt.subplots(figsize=(10, 8))

sns.countplot(x="Churn", data=df_telco)

plt.title('Quantidade de observações pela classe Alvo')
plt.xlabel('Classe Alvo')
plt.ylabel('Quantidade de Observações')
plt.show()
```

```python
# Separando os vetores de atributos da variável dependente
X = df_telco.drop('Churn',axis=1)
Y = df_telco['Churn']

# Separando os conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X,Y, random_state = 12)
```

## 4.0 Models

```python
# Avaliação de modelos de classificação

def resultados_classificacao(modelo, nome_do_modelo ):
    y_pred = np.nan
    y_pred = modelo.predict(X_test)
    y_pred_proba = modelo.predict_proba(X_test)[:, 1]
        
    acuracia  =  accuracy_score(y_test,y_pred)
    f1        =  f1_score(y_test, y_pred) 
    precisao  =  precision_score(y_test, y_pred)
    recall    =  recall_score(y_test,y_pred)

    print('Resultados:')
    print('Acuracia                     : ', round(acuracia*100, 2) , '%'  )
    print('F1 Score                     : ', round(f1*100, 2)        , '%'  )
    print('Precisão (Precision)         : ', round(precisao*100, 2)  , '%' )
    print('Recall                       : ', round(recall*100 , 2)   , '%'  )

    fig, axes = plt.subplots(1, 2, figsize=(20, 8))

    # Matriz de Confusão
    disp = plot_confusion_matrix(modelo, X_test, y_test, cmap=plt.cm.Blues, ax = axes[0])

    [fpr, tpr, thr] = roc_curve(y_test, y_pred_proba)

    idx = np.min(np.where(tpr > 0.95)) # index of the first threshold for which the sensibility > 0.95

    axes[1].plot(fpr, tpr, color='coral', label='ROC curve (area = %0.3f)' % auc(fpr, tpr))
    axes[1].plot([0, 1], [0, 1], 'k--')
    axes[1].plot([0,fpr[idx]], [tpr[idx],tpr[idx]], 'k--', color='blue')
    axes[1].plot([fpr[idx],fpr[idx]], [0,tpr[idx]], 'k--', color='blue')

    axes[1].set_xlabel('False Positive Rate (1 - specificity)', fontsize=14)
    axes[1].set_ylabel('True Positive Rate (recall)', fontsize=14)
    axes[1].set_title('Receiver operating characteristic (ROC) curve')
    axes[1].legend(loc="lower right");
```

## 5.1 Regressão Logística

```python
# Regressão Logística sem padronização dos dados
print("Modelo 1.1 - Regressão Logística sem normalização\n") 

# Cria o objeto com o modelo do SKLearn
LR = LogisticRegression()

LR.fit(X_train,y_train)
y_pred = LR.predict(X_test)

# Verifica os resultados
resultados_classificacao(LR, 'Logistic Regression')
```

```python
# Regressão Logística com MinMaxScaler
print("Modelo 1.2 - Regressão Logística com MinMaxScaler\n") 
minmax_scaler = MinMaxScaler()

X_train, X_test, y_train, y_test = train_test_split(X,Y, random_state = 12)

X_train = minmax_scaler.fit_transform(X_train)
X_test = minmax_scaler.transform(X_test)

LR.fit(X_train,y_train)
y_pred = LR.predict(X_test)

# Verifica os resultados
resultados_classificacao(LR, 'Logistic Regression')
```

```python
# Regressão Logística com StandardScaler
print("Modelo 1.3 - Regressão Logística com StandardScaler\n") 
std_scaler = StandardScaler()

X_train, X_test, y_train, y_test = train_test_split(X,Y, random_state = 12)

X_train = std_scaler.fit_transform(X_train)
X_test = std_scaler.transform(X_test)

LR.fit(X_train,y_train)
y_pred = LR.predict(X_test)

# Verifica os resultados
resultados_classificacao(LR, 'Logistic Regression')
```

## 4.2 Árvore de Decisão

```python
# Decision Tree Classifier
print("Modelo 2 - Árvore de Decisão\n") 

# Cria o objeto com o modelo do SKLearn
decision_tree = DecisionTreeClassifier(max_depth=4)

# Treina o modelo para os dados de Treino
decision_tree.fit(X_train, y_train)

# Verifica os resultados
resultados_classificacao(decision_tree, 'Decision Tree')
```

```python
plt.figure(figsize=(20,12))
tree.plot_tree(decision_tree,filled=True)
```

## 4.3 Random Forest

```python
# Decision Tree Classifier
print("Modelo 3 - Floresta Aleatória\n") 

# Cria o objeto com o modelo do SKLearn
random_forest = RandomForestClassifier()

# Treina o modelo para os dados de Treino
random_forest.fit(X_train, y_train)

# Verifica os resultados
resultados_classificacao(random_forest, 'Floresta Aleatória')
```

## 4.4 Naive Bayes

```python
# Naive Bayes
print("Modelo 4 - Naive Bayes\n") 

# Cria o objeto com o modelo do SKLearn
naive_bayes = GaussianNB()

# Treina o modelo para os dados de Treino
naive_bayes.fit(X_train, y_train)

# Verifica os resultados
resultados_classificacao(naive_bayes, 'Naive Bayes')
```

## 4.5 Support Vector Machine

```python
# Support Vector Classifier
print("Modelo 5 -Support Vector Classifier \n") 

# Cria o objeto com o modelo do SKLearn
svc_model = SVC(probability=True)

# Treina o modelo para os dados de Treino
svc_model.fit(X_train, y_train)

# Verifica os resultados
resultados_classificacao(svc_model, 'Support Vector Classifier')
```
